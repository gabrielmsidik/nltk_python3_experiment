## Prepare Corpus

**Load books from files**

```python
book_filenames = sorted(glob.glob("data/iphone.txt"))
```

```python
print("Found books:")
print(book_filenames)
```

**Combine the books into one string**

```python
corpus_raw = u""
print("corpus_raw", corpus_raw)
for book_filename in book_filenames:
    print("Reading '{0}'...".format(book_filename))
    with codecs.open(book_filename, "r") as book_file:
        corpus_raw += book_file.read()
    print("Corpus is now {0} characters long".format(len(corpus_raw)))
    print()
```

**Split the corpus into sentences**

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
raw_sentences = tokenizer.tokenize(corpus_raw)
# raw_sentences = raw_sentences[0].split("\n")
print(raw_sentences)
```

```python
#convert into a list of words
#remove unnecessary,, split into words, no hyphens
#list of words

def sentence_to_wordlist(raw):
    clean = re.sub("[^a-zA-Z]"," ", raw)
    words = clean.split()
    return words
```

```python
#sentence where each word is tokenized
sentences = []
for raw_sentence in raw_sentences:
    if len(raw_sentence) > 0:
        sentences.append(sentence_to_wordlist(raw_sentence))
```

```python
# print(raw_sentence)
# print(sentence_to_wordlist(raw_sentences))
```

```python
token_count = sum([len(sentence) for sentence in sentences])
print("The song corpus contains {0:,} tokens".format(token_count))
```

## Train Word2Vec

```python
#ONCE we have vectors
#step 3 - build model
#3 main tasks that vectors help with
#DISTANCE, SIMILARITY, RANKING

# Dimensionality of the resulting word vectors.
#more dimensions, more computationally expensive to train
#but also more accurate
#more dimensions = more generalized
num_features = 300
# Minimum word count threshold.
min_word_count = 1

# Number of threads to run in parallel.
#more workers, faster we train
num_workers = multiprocessing.cpu_count()

# Context window length.
context_size = 7

# Downsample setting for frequent words.
#0 - 1e-5 is good for this
downsampling = 1e-3

# Seed for the RNG, to make the results reproducible.
#random number generator
#deterministic, good for debugging
seed = 1
```

```python
song2vec = w2v.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling
)
```

```python
song2vec.build_vocab(sentences)
```

```python
print("Word2Vec vocabulary length:", len(song2vec.wv.vocab))
```

**Start training, this might take a minute or two...**

```python
song2vec.train(sentences, total_examples=song2vec.corpus_count, epochs=10)
```

**Save to file, can be useful later**

```python
if not os.path.exists("trained"):
    os.makedirs("trained")
```

```python
song2vec.save(os.path.join("trained", "iphone2vec.w2v"))
```

## Explore the trained model.

```python
song2vec = w2v.Word2Vec.load(os.path.join("trained", "iphone2vec.w2v"))
```

### Compress the word vectors into 2D space and plot them

```python
#my video - how to visualize a dataset easily
tsne = sklearn.manifold.TSNE(n_components=2, random_state=0)
```

```python
all_word_vectors_matrix = song2vec.wv.vectors
```

**Train t-SNE, this could take a minute or two...**

```python
all_word_vectors_matrix_2d = tsne.fit_transform(all_word_vectors_matrix)
```

**Plot the big picture**

```python
points = pd.DataFrame(
    [
        (word, coords[0], coords[1])
        for word, coords in [
            (word, all_word_vectors_matrix_2d[song2vec.wv.vocab[word].index])
            for word in song2vec.wv.vocab
        ]
    ],
    columns=["word", "x", "y"]
)
```

```python
points.head(10)
```

```python
sns.set_context("poster")
```

```python
points.plot.scatter("x", "y", s=10, figsize=(20, 12))
```

**Zoom in to some interesting places**

```python
def plot_region(x_bounds, y_bounds):
    slice = points[
        (x_bounds[0] <= points.x) &
        (points.x <= x_bounds[1]) & 
        (y_bounds[0] <= points.y) &
        (points.y <= y_bounds[1])
    ]
    
    ax = slice.plot.scatter("x", "y", s=35, c=None, figsize=(10, 8))
    for i, point in slice.iterrows():
        ax.text(point.x + 0.005, point.y + 0.005, point.word, fontsize=11)
```

**Found the House Mottos!**

```python
plot_region(x_bounds=(0, 2), y_bounds=(0, 2))
```

**Can't find anything particularly interesting anymore :(**

```python
plot_region(x_bounds=(0, 5), y_bounds=(0, 5))
```

### Explore semantic similarities between book characters

**Words closest to the given word**

```python
song2vec.wv.most_similar("power")
```

```python
song2vec.wv.most_similar("Internet")
```

```python
song2vec.wv.most_similar("music")
```

**Linear relationships between word pairs**

```python
def nearest_similarity_cosmul(start1, end1, end2):
    similarities = song2vec.wv.most_similar_cosmul(
        positive=[end2, start1],
        negative=[end1]
    )
    start2 = similarities[0][0]
    print("{start1} is related to {end1}, as {start2} is related to {end2}".format(**locals()))
    return start2
```

```python
nearest_similarity_cosmul("screen", "brightness", "loud")
```
